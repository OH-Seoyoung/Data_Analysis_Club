```python
# Import Package
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [1,2,3]
plt.plot(x , y)
plt.show()
```

```python
import numpy as np

x = np.linspace(0 , np.pi * 10 , 500)
fig, axes = plt.subplots(2 , 1)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
plt.show()
```

```python
import numpy as np

x = np.linspace(0 , np.pi * 10 , 10)
fig, axes = plt.subplots(2 , 1)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
plt.show()
```

```python
x = np.arange(-9,10)
y1 = x**2
y2 = -x
plt.plot(x,y1,color = 'red')
plt.plot(x,y2,color = 'blue')
plt.show()
```

```python
x = np.arange(-9,10)
plt.bar(x,x**2)
plt.title("My graph")
plt.xlabel("X")
plt.ylabel("X**2")
plt.show()
```
