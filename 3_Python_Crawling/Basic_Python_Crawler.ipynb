```python
def p(x,y):
    result = print("%s 과 %s 의 더한 값은 %s 입니다" %(x,y,x+y))
    return result
```

```python
p(3,4)
```

```python
def p(*x):
    hap = 0
    for i in x:
        hap+=i
    return ('입력하신 숫자의 합은 %s 입니다' %hap) 
```

```python
print(p(1,3,5))
```

```python
print(p(1,2,3,4,5,6,7))
```
